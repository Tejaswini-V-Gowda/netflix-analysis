"""

Objective:
This project aims to perform a comprehensive exploratory data analysis (EDA) on the Netflix Titles Dataset. The primary goal is to uncover key trends related to content types, release patterns, popular genres, and the geographical distribution of content. Methodology:
1.  Data Understanding & Cleaning: Loaded the dataset, checked for missing values, and performed necessary cleaning steps, including converting data types and filling null values.
2.  Feature Engineering: Created new features such as `release_decade`, `is_movie`, and a consolidated `country_group` to simplify analysis.
3.  Exploratory Data Analysis (EDA): Utilized various visualizations to explore the data.
    -   **Pie Chart:** To show the distribution of Movies vs. TV Shows.
    -   **Line Plot:** To visualize the trend of content added per year.
    -   **Bar Plot:** To identify the top 10 most popular genres.
    -   **Heatmap:** To compare content volume by country and content type.
4.  Insight Generation: Interpreted the visualizations to derive meaningful insights about Netflix's content strategy.
5.  Bonus: Trend Prediction: Employed a simple forecasting model to predict content additions for the next two years based on historical data.

Key Findings:

-   The number of titles added to the platform has shown explosive growth, particularly in recent years, indicating an aggressive content strategy.

-   The United States is the leading contributor of content, followed by India and the United Kingdom.

Libraries Used:
-   pandas

```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load dataset
df = pd.read_csv(r"C:\Users\tejas\skybricks\task4\data\netflix_titles.csv")

# Step 3: Basic info
print("Dataset Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nData Types:\n", df.dtypes)

# Step 4: First 5 rows
print("\nSample Data:\n", df.head())

# Step 5: Missing values check
print("\nMissing Values:\n", df.isnull().sum())
```

## **Step 1: Data Cleaning**

```python
# Convert date_added to datetime
df["date_added"] = pd.to_datetime(df["date_added"], errors="coerce")

# Fill missing categorical values with "Unknown"
# Instead of inplace=True (which causes warning)
df["director"] = df["director"].fillna("Unknown")
df["cast"] = df["cast"].fillna("Unknown")
df["country"] = df["country"].fillna("Unknown")
df["rating"] = df["rating"].fillna("Unknown")
df["duration"] = df["duration"].fillna("Unknown")


# Create new feature: release_decade
df["release_decade"] = (df["release_year"] // 10) * 10

# Create new feature: is_movie (1 for Movie, 0 for TV Show)
df["is_movie"] = df["type"].apply(lambda x: 1 if x == "Movie" else 0)

# Verify cleaning
print("\nMissing Values after cleaning:\n", df.isnull().sum())
print("\nData Types after cleaning:\n", df.dtypes)
print("\nSample Data after cleaning:\n", df.head())
```

```python
# Find top 5 countries with the most Netflix titles
top_countries = df['country'].value_counts().head(5)

print("Top 5 Countries with most content:")
print(top_countries)

# Create a new column to group countries
df['country_group'] = df['country'].apply(lambda x: x if x in top_countries.index else 'Other')

# Preview new column
df[['title', 'country', 'country_group']].head(10)
```

```python
import matplotlib.pyplot as plt

# Count of Movies vs TV Shows
type_counts = df['type'].value_counts()

# Plot
plt.figure(figsize=(6,6))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#99ff99'])
plt.title("Distribution of Movies vs TV Shows on Netflix")
plt.show()
```

```python
import seaborn as sns

# Count of content by release year
year_counts = df['release_year'].value_counts().sort_index()

# Plot
plt.figure(figsize=(12,6))
sns.lineplot(x=year_counts.index, y=year_counts.values, marker='o', color="red")
plt.title("Netflix Content Released Per Year", fontsize=14, fontweight='bold')
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.grid(True)
plt.show()
```

```python
# Create a list to store all individual genres
genres = []
for genre_list in df['listed_in'].str.split(','):
    for genre in genre_list:
        genres.append(genre.strip()) # .strip() removes leading/trailing whitespace

# Create a Series from the list of genres and get the top 10
top_10_genres = pd.Series(genres).value_counts().head(10)

# Create the genre_df dataframe for plotting
genre_df = top_10_genres.reset_index()
genre_df.columns = ['Genre', 'Count']

# Now, the plotting code will work without the NameError
plt.figure(figsize=(12, 6))
sns.barplot(
    data=genre_df,
    x="Count",
    y="Genre",
    hue="Genre",
    dodge=False,
    palette="viridis",
    legend=False
)
plt.title("Top 10 Genres on Netflix", fontsize=14, fontweight="bold")
plt.xlabel("Number of Titles")
plt.ylabel("Genre")
plt.show()
```

```python
# Prepare data for heatmap: Country vs Content Type
country_content_counts = df.groupby(['country_group', 'type']).size().unstack(fill_value=0)

# Reorder columns for better visualization
country_content_counts = country_content_counts[['Movie', 'TV Show']]

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(country_content_counts, annot=True, fmt='d', cmap='YlGnBu', linewidths=0.5, linecolor='white')
plt.title("Content Volume by Top Countries (Movie vs. TV Show)", fontsize=14, fontweight='bold')
plt.xlabel("Content Type")
plt.ylabel("Country Group")
plt.show()

# After the heatmap, we will provide the final deliverables
# This includes the summary of insights and the bonus trend prediction.
```

```python
# Create a new dataframe with content added by year from the date_added column
df_added_by_year = df.dropna(subset=['date_added'])
yearly_content_added = df_added_by_year['date_added'].dt.year.value_counts().sort_index().reset_index()
yearly_content_added.columns = ['year', 'count']

# Focus on the last 5 years for a more relevant trend
recent_years = yearly_content_added[yearly_content_added['year'] >= 2018]
print("\nRecent Yearly Content Additions:\n", recent_years)

# Simple linear projection to predict next 2 years
# We'll use the average yearly increase from the recent data
average_increase = np.mean(np.diff(recent_years['count']))
last_year_count = recent_years['count'].iloc[-1]
last_year = recent_years['year'].iloc[-1]

# Predict for the next two years
next_year_pred = last_year_count + average_increase
two_years_pred = next_year_pred + average_increase

print(f"\nPredicted Titles for {last_year + 1}: {int(next_year_pred)} titles")
print(f"Predicted Titles for {last_year + 2}: {int(two_years_pred)} titles")

# You can also visualize this prediction
prediction_years = pd.DataFrame({
    'year': [last_year + 1, last_year + 2],
    'count': [next_year_pred, two_years_pred]
})

all_data = pd.concat([yearly_content_added, prediction_years])

plt.figure(figsize=(12, 6))
sns.lineplot(data=all_data, x='year', y='count', marker='o', color='blue', label='Actual & Predicted Content')
plt.axvline(x=last_year, color='red', linestyle='--', label='Prediction Start')
plt.title("Content Addition Trend with 2-Year Forecast", fontsize=14, fontweight='bold')
plt.xlabel("Year")
plt.ylabel("Number of Titles Added")
plt.legend()
plt.grid(True)
plt.show()
```
